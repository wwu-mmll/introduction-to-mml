# Python Packages

- Python packages = Collections of modules and functions
- Reuse code written by other developers -> saving you time and effort
- Two types of packages
  - Buildin (e.g. time, os)
  - Not-Buildin (e.g. numpy, pandas) -> Installation needed (in notebook `!pip install numpy`, in terminal `pip install numpy`)



### Example 1: Getting the Current Time

The `time` package provides a function called `time()` that returns the number of seconds since the Unix epoch. You can use this function to get the current time.

```python
import time

current_time = time.time()
print(current_time)  # Output: 1624979806.0477715
```

The `time()` function returns a floating-point number representing the current time in seconds.

**Task 1:** Sum all integers from 0 to 10000 in a for loop and use the `time()` function to find out how long the execution takes.


````python
import time

start_time = time.time()
numbers = []
for i in range(10000):
    numbers.append(i)
list_sum = sum(numbers)
end_time = time.time()
execution_time = end_time - start_time
print(execution_time)
````

### Example 2: Delaying Execution

The `time` package includes a function called `sleep()` that allows you to pause the execution of your program for a specified number of seconds. Here's an example:

```python
import time

print('Before sleep')
time.sleep(3)
print('After sleep')
```

Output:
```
Before sleep
<waits for 3 seconds>
After sleep
```

In this example, the program pauses for 3 seconds after the "Before sleep" message is printed, and then continues execution after the sleep period.


### Using from... import

If you only want to use one or two functions (e.g. `time()` and `sleep()`) you can only import them using

```python
from time import time, sleep
```

This is especially useful if the desired function is very deep in the package tree

```python
from package_xy.module_z.submodule1 import nice_function
```

This way you do not have to write `package_xy.module_z.submodule1.nice_function()` if you want to use this function
