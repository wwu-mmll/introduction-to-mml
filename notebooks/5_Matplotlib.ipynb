# Matplotlib

Matplotlib is a widely used data visualization library that allows you to create various types of plots.

By convention, `matplotlib.pyplot` is commonly imported as `plt` to make it easier to reference the library's functions and methods.

```python
import matplotlib.pyplot as plt
```

## 1. Basic Plotting with Matplotlib

The most commonly used functions is `plot()`, which is used to create line plots.

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

plt.plot(x, y)
plt.show()
```


````python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

plt.plot(x, y)
plt.show()
````


In this example, we create a simple line plot by passing the x-coordinates and y-coordinates as arguments to the `plot()` function.

The `show()` function is then called to display the plot.

## 2. Customizing Plots

Matplotlib allows you to modify the appearance of lines, markers, colors, labels, titles, axes... Lets use it!


````python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

plt.plot(x, y, color='red', linewidth=3)
plt.title('Important plot')
plt.xlabel('Input Data [x]')
plt.ylabel('Output Data [y]')
plt.show()
````


## 3. Popular plotting functions

By replacing `plt.plot()` which is creating a lineplot, other kinds of plots can be created:

- `plt.scatter()` creates a scatter plot


````python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

plt.scatter(x, y)
plt.show()
````


- `plt.hist()` plots a histogram

````python
import matplotlib.pyplot as plt

x = [1, 2, 2, 3, 3, 3, 4, 4, 5]

plt.hist(x)
plt.show()
````


- `plt.imshow()` plots images - i.e. two dimensional arrays with pixel values


````python
import numpy as np
import matplotlib.pyplot as plt

array = np.random.rand(100, 100)

plt.imshow(array)
plt.colorbar()
plt.show()
````

## 4. Multiple Plots

In Matplotlib, there exist many ways to create multiple plots within a single figure. Here I will show the **most convenient way using subfigures**. In older stackoverflow posts you will see other recommendations since subfigures are a newish feature of Matplotlib.

We can use `subfigures()` to divide a figure into multiple subfigures arranged in a grid.

```python
import matplotlib.pyplot as plt

# Create a figure with constrained layout (automatic layout adjustments)
fig = plt.figure(constrained_layout=True)
subfigs = fig.subfigures(2, 3)  # nrows=2, ncols=3
```

Using `subplots()` you can create a 2D array of subplots (typically named `axs`) within a subfigure (or figure). With `axs[0, 2].plot(x, y)` instead of `plt.plot(x, y)` you can specify you would plot in subplot in row 0 and column 3.

If either `nrows` or `ncols` in `subplots(nrows, ncols)` is 1, `axs[3]` instead of `axs[0, 3]` or `axs[3, 0]` is used.

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

# Create a figure with constrained layout (automatic layout adjustments)
fig = plt.figure(constrained_layout=True)
subfigs = fig.subfigures(2, 3)  # nrows=2, ncols=3
# Subplots in one subfigure
axs = subfigs[0, 0].subplots(2, 2)  # nrows=2, ncols=2
axs[0, 1].plot(x, y)
axs[1, 0].scatter(x, y)
# Subplots in another subfigure
axs = subfigs[0, 1].subplots(1, 2)  # nrows=1, ncols=2
axs[0].plot(x, y)
# More plotting...

plt.show()
```
Here is an typical example which uses all learned concepts


````python
import numpy as np
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]
array = np.random.rand(100, 100)

# Create a figure with constrained layout (automatic layout adjustments)
fig = plt.figure(constrained_layout=True)
subfigs = fig.subfigures(2, 1)  # nrows=2, ncols=1

# Create subplots within first subfigure
axs = subfigs[0].subplots(1, 2)  # nrows=1, ncols=2
# Plot in first subplot
axs[0].plot(x, y)
# Scatter in second subplot
axs[1].scatter(x, y)
# Customize subplots as needed
axs[0].set_title('Plot')
axs[1].set_title('Scatter')
axs[1].set_xlabel('X')
axs[1].set_ylabel('Y')
axs[1].set_xlim(0, 10)

# Create single subplots within second subfigure
ax = subfigs[1].subplots(1)
# Plot image in subplot
ax.imshow(array)

plt.show()
````

## 5. Save to file

Figures can be easily saved as .png-, .pdf- or .svg-files using `plt.savefig('path/to/example.png')`.

```python
import numpy as np
import matplotlib.pyplot as plt

array = np.random.rand(100, 100)

plt.imshow(array)
plt.colorbar()
plt.savefig('path/to/example.png')
```
**Important warning**: If `plt.show()` is called before `plt.savefig()`, the saved file will be empty.

So, the following will show the plot but save an empty figure.

```python
import numpy as np
import matplotlib.pyplot as plt

array = np.random.rand(100, 100)

plt.imshow(array)
plt.colorbar()
plt.show()
plt.savefig('path/to/example.png')
```

## 6. Plotting is a case for ChatGPT

Plotting sophisticated stuff can take a lot of special Matplotlib functions you have to Google. Unlike in most coding tasks, in plotting ChatGPT can not really mess up without you directly noticing (via looking at the plot).

Thats why you should **just ask ChatGPT to do your sophisticated plot**!

Here is the code ChatGPT outputed when I asked it:

"Write python code which displays a donut as a surface in matplotlib!"

````python
import numpy as np
import matplotlib.pyplot as plt

# Create a grid of theta and phi values
theta = np.linspace(0, 2 * np.pi, 100)
phi = np.linspace(0, 2 * np.pi, 100)
theta, phi = np.meshgrid(theta, phi)

# Define the parameters for the donut
R = 1  # Outer radius
r = 0.5  # Inner radius

# Compute the Cartesian coordinates for each point on the surface
x = (R + r * np.cos(phi)) * np.cos(theta)
y = (R + r * np.cos(phi)) * np.sin(theta)
z = r * np.sin(phi)

# Create a 3D plot and display the donut surface
fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(x, y, z, cmap='viridis')

# Set labels and title
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('Donut Surface')

# Set aspect ratio and display the plot
ax.set_box_aspect([1, 1, .3])  # Equal aspect ratio for all axes
plt.show()
````
